## Criando Dimensões

```python
# pasta base onde vão ficar as dimensões
BASE_DIR = Path(r"C:/Users/gui-b/Teste-tecnico/teste-tecnico/bases/dimensao")

def limpar_descricao_cid(s: str) -> str:
    """Removedno os códigos do início (ICD como A00, A00.1) para manter somente a descrição."""
    if pd.isna(s):
        return s
    t = str(s)
    # removendo os códigos
    t = re.sub(r'^\s*[A-Z]\d{2,3}(?:\.[0-9A-Z]{1,2})?\s*', '', t)
    # removendo os códigos numericos que estão no começo: 0, 20...
    t = re.sub(r'^\s*\d+(?:\.\d+)?\s*', '', t)
    # normalizando os espaços existentes
    t = re.sub(r'\s{2,}', ' ', t).strip(" -–—\t")
    return t

def make_dim(df_fev: pd.DataFrame, code_col: str, desc_col: str, pasta_nome: str, prefix: str,
             desc_cleaner=None) -> pd.DataFrame:
    dim = (df_fev[[code_col, desc_col]]
           .rename(columns={code_col: f"CD_{prefix}", desc_col: f"DS_{prefix}"})
           .dropna(subset=[f"CD_{prefix}", f"DS_{prefix}"])
          )

    # normalizando: para string
    dim[f"CD_{prefix}"] = dim[f"CD_{prefix}"].astype(str).str.strip()
    dim[f"DS_{prefix}"] = dim[f"DS_{prefix}"].astype(str).str.strip()

    
    if desc_cleaner is not None:
        dim[f"DS_{prefix}"] = dim[f"DS_{prefix}"].map(desc_cleaner)

    # removendo duplicatas e ordenando
    dim = dim.drop_duplicates()
    dim = dim.sort_values([f"CD_{prefix}", f"DS_{prefix}"], kind="stable")

    out_dir = BASE_DIR / pasta_nome
    out_dir.mkdir(parents=True, exist_ok=True)
    out_path = out_dir / f"dim_{prefix.lower()}.csv"
    dim.to_csv(out_path, index=False, encoding="utf-8-sig")
    print(f"✅ Dimensão {prefix} gerada com {len(dim)} linhas -> {out_path}")
    return dim


dim_especie  = make_dim(df_fev, "Espécie",   "Espécie.1",   "Especie",  "ESPECIE")
dim_cid      = make_dim(df_fev, "CID",       "CID.1",       "CID",      "CID",      desc_cleaner=limpar_descricao_cid)
dim_despacho = make_dim(df_fev, "Despacho",  "Despacho.1",  "Despacho", "DESPACHO")
```

## CONSOLIDANDO BASE FINAL

```python
from pathlib import Path
import pandas as pd
import re

base_dir = Path(r"C:/Users/gui-b/Teste-tecnico/teste-tecnico/bases")
saida_dir = base_dir / "tratadas"
saida_dir.mkdir(parents=True, exist_ok=True)

jan_name = "CONCEDIDOS_DADOS_ABERTOS_JANEIRO+2025.xlsx"

def normalize_columns(cols):
    # 1) strip + colapsa espaços internos
    norm = [re.sub(r"\s+", " ", str(c).strip()) for c in cols]
    # 2) deduplica adicionando .1, .2...
    seen = {}
    out = []
    for c in norm:
        if c not in seen:
            seen[c] = 0
            out.append(c)
        else:
            seen[c] += 1
            out.append(f"{c}.{seen[c]}")
    return out

dfs = []
ordem_columns = None 

for f in sorted(base_dir.glob("*.xlsx")):
    header = 0 if f.name == jan_name else 1
    print(f"Lendo {f.name} (header={header})")
    df = pd.read_excel(f, header=header, engine="calamine")

    # Normalizando nomes
    df.columns = normalize_columns(df.columns)

    # Padronizando algumas variaveis que mudam de nome
    ren = {
        "Qt SM RMI": "Qt SM RMI",
        "Sexo.": "Sexo.",  #mantendo esse por garantia
        
    }
    df.rename(columns=ren, inplace=True)

    # ordem a partir de janeiro
    if f.name == jan_name:
        ordem_columns = list(df.columns)

    dfs.append(df)


df_consolidado = pd.concat(dfs, ignore_index=True, sort=False)

# Reordenando seguindo as ordens
if ordem_columns is not None:
    other_cols = [c for c in df_consolidado.columns if c not in ordem_columns]
    df_consolidado = df_consolidado[ordem_columns + other_cols]

saida_csv = saida_dir / "BASE_CONSOLIDADA.csv"
df_consolidado.to_csv(saida_csv, index=False, encoding="utf-8-sig")

print(f"OK -> {saida_csv} | linhas: {len(df_consolidado)} | colunas: {len(df_consolidado.columns)}")
```
